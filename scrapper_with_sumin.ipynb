```python
import requests
from bs4 import BeautifulSoup

def get_goods_name_link(bs4_tags:list)->list:
    result = []
    for bs4_tag in bs4_tags:
        result.append(bs4_tag["href"])
    return result

def get_goods_name_tags(URL):
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser")
    goods_name = soup.select("div.goods_info div.goods_name a")
    return goods_name
    

def get_bookclub_books_tag():
    URL = f"http://bookclub.yes24.com/BookClub/BcCateSub?OrderBy=UPDATE&FilterBy=&pageNo=1&title=%EB%B6%84%EC%95%BC%20%EC%84%A0%ED%83%9D"
    goods_name_tags = get_goods_name_tags(URL)
    first_page_cnt = len(goods_name_tags) 
    i = 2
    while True:
        URL = f"http://bookclub.yes24.com/BookClub/BcCateSub?OrderBy=UPDATE&FilterBy=&pageNo={i}&title=%EB%B6%84%EC%95%BC%20%EC%84%A0%ED%83%9D"
        current_page_name_tag = get_goods_name_tags(URL)
        goods_name_tags += current_page_name_tag
        if len(current_page_name_tag) != first_page_cnt:
            break
        i += 1
        print(i)
    return goods_name_tags

def init():
    res = get_bookclub_books_tag()
    res =get_goods_name_link(res)
    return res
    
```

```python
test_result = init()    
```

```python
#[[],[],[],[],[]]
def get_book_info(bookURL,book_info_values:list):
    platform_name = 'yes24bookclub'
    response = requests.get(bookURL)
    soup = BeautifulSoup(response.text, "html.parser")
    book_info_tag = soup.select("div.bcDetail_info")[0]
    
    title = book_info_tag.find('h2').text
    author = book_info_tag.find('span',{'bcDetail_auth'}).text
    pub = book_info_tag.find('span',{'bcDetail_pub'}).text
    pub_date = book_info_tag.find('span',{'bcDetail_date'}).text
    
    book_info_value = [title, author, pub, pub_date,platform_name]
    for i in range(len(book_info_value)):
        book_info_values[i].append(book_info_value[i])      
    return book_info_values
```
