```python
import requests
from bs4 import BeautifulSoup

pip install beautifulsoup4

URL = "http://bookclub.yes24.com/BookClub/BcCateSub?OrderBy=UPDATE&FilterBy=&pageNo=1&title=%EB%B6%84%EC%95%BC%20%EC%84%A0%ED%83%9D"

def get_hrefs(URL):
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser")
    goods_name = soup.select("div.goods_info div.goods_name a")
    return goods_name
    

def amolgang(start, end):
    total_hrefs =[]
    for i in range(start,end+1):
        URL = f"http://bookclub.yes24.com/BookClub/BcCateSub?OrderBy=UPDATE&FilterBy=&pageNo={i}&title=%EB%B6%84%EC%95%BC%20%EC%84%A0%ED%83%9D"
        total_hrefs += get_hrefs(URL)
    return total_hrefs

def init():
    res = amolgang(1,3)
    return res
    
```
